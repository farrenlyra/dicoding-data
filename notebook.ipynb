### Cleaning Data

1. **Menghapus Tabel yang Tidak Diperlukan**

Kita tidak memerlukan data dengan resolusi jam (hourly), cukup menggunakan data harian (daily) karena pertanyaan bisnis lebih fokus pada perbandingan pola antar hari, cuaca, dan musim.

```python
del hour_df
```

2. **Menghapus Kolom yang Tidak Diperlukan**

Kolom yang tidak berkaitan langsung dengan pertanyaan bisnis akan dihapus. Berdasarkan pertanyaan bisnis, kita tidak membutuhkan:

**instant**: Indeks catatan ini tidak memberikan informasi yang relevan.

**windspeed**: Tidak ada pertanyaan yang berfokus pada pengaruh kecepatan angin terhadap jumlah penyewaan.

```python
drop_col = ['instant', 'windspeed']
for i in day_df.columns:
    if i in drop_col:
        day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

3. **Mengubah Judul Kolom**

Kita akan mengubah nama beberapa kolom agar lebih mudah dipahami dan konsisten dengan pertanyaan bisnis. Selain itu, kita akan mengubah tipe data yang lebih sesuai, misalnya, mengubah angka musiman atau hari dalam seminggu menjadi bentuk kategori yang lebih mudah dibaca.

**Mengganti Nama Kolom**

```python
day_df.rename(columns={
    'dteday': 'date',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_condition',
    'cnt': 'total_count'
}, inplace=True)
day_df.head()
```

**Mengubah Angka menjadi Label Kategori**

Agar lebih mudah dianalisis, kita akan mengonversi nilai angka pada beberapa kolom menjadi deskripsi kategori.

```python
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})

day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})

day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})

day_df['weather_condition'] = day_df['weather_condition'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

day_df.head()
```

```python
day_df.info()
```

## Exploratory Data Analysis (EDA)

1. Kelompokkan penyewa sepeda (casual dan registered) berdasarkan bulan

```python
day_df.groupby(by='month').agg({
    'total_count': ['max', 'min', 'mean', 'sum']
})
```

Kesimpulan: Bulan dengan rata-rata penyewaan terbesar adalah bulan Juni. Sementara itu, bulan Januari memiliki rata-rata terkecil.

2. Kelompokkan penyewa sepeda berdasarkan kondisi cuaca

```python
day_df.groupby(by='weather_condition').agg({
    'total_count': ['max', 'min', 'mean', 'sum']
})
```

Kesimpulan: Para penyewa cenderung menyewa sepeda pada saat cuaca cerah atau mendung. Penyewaan jauh lebih sedikit saat salju turun atau hujan ringan.

3. Kelompokkan penyewa sepeda berdasarkan hari libur

```python
day_df.groupby(by='holiday').agg({
    'total_count': ['max', 'min', 'mean', 'sum']
})
```

Kesimpulan: Penyewa lebih banyak menyewa sepeda saat bukan hari libur dibandingkan pada hari-hari libur.

4. Bandingkan penyewaan sepeda pada hari kerja (weekday) dan akhir pekan (weekend)

```python
day_df.groupby(by='weekday').agg({
    'total_count': ['max', 'min', 'mean']
})
```

Kesimpulan: Hari Jumat memiliki rata-rata penyewaan tertinggi, diikuti oleh Kamis dan Sabtu. Minggu dan Senin memiliki rata-rata penyewaan yang lebih rendah.

5. Kelompokkan penyewa sepeda berdasarkan hari kerja (workingday)

```python
day_df.groupby(by='workingday').agg({
    'total_count': ['max', 'min', 'mean']
})
```

Kesimpulan: Penyewaan sepeda pada hari kerja lebih tinggi dibandingkan dengan hari non-kerja, namun perbedaannya tidak terlalu signifikan.

6. Kelompokkan penyewa sepeda berdasarkan musim (season)

```python
day_df.groupby(by='season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'total_count': ['max', 'min', 'mean']
})
```

Kesimpulan: Penyewaan sepeda tertinggi terjadi pada musim gugur (Fall) dengan rata-rata tertinggi baik dari penyewa casual maupun registered. Penyewaan terendah ada di musim semi (Spring).

7. Kelompokkan suhu (temp), suhu persepsi (atemp), dan kelembaban (hum) berdasarkan musim

```python
day_df.groupby(by='season').agg({
    'temp': ['max', 'min', 'mean'],
    'atemp': ['max', 'min', 'mean'],
    'hum': ['max', 'min', 'mean']
})
```

Kesimpulan:

*   Suhu dan suhu persepsi tertinggi terjadi pada musim gugur (Fall).
*   Kelembaban tertinggi ada pada musim dingin (Winter).

8. Hubungan antara casual, registered, dan total_count

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
correlation_matrix = day_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', ax=ax)
plt.show()
```

Kesimpulan:

* **atemp dan temp sangat berkorelasi (0.99)**. Ini berarti temperatur yang dirasakan (atemp) memiliki hubungan yang hampir sempurna dengan suhu aktual (temp), yang masuk akal karena keduanya sangat berhubungan.

* **hum memiliki korelasi lemah dengan temp dan atemp (0.13 dan 0.14)**. Ini menunjukkan bahwa kelembapan (hum) tidak memiliki hubungan yang signifikan dengan suhu ataupun temperatur yang dirasakan.

* **casual cukup berkorelasi dengan temp dan atemp (0.54)**. Ini berarti bahwa jumlah penyewa sepeda kasual (casual) meningkat seiring dengan meningkatnya suhu, dan orang cenderung menyewa sepeda lebih banyak ketika cuaca lebih hangat.

* **casual memiliki sedikit korelasi negatif dengan hum (-0.08)**. Ini menunjukkan bahwa semakin tinggi kelembapan, sedikit lebih rendah jumlah penyewa kasual, meskipun pengaruhnya relatif kecil.

* **registered memiliki pola yang mirip dengan casual, dengan korelasi moderat terhadap temp dan atemp (0.54 dan 0.54), namun memiliki korelasi yang lebih kuat dengan casual (0.40)**. Ini menunjukkan bahwa faktor cuaca juga memengaruhi penyewa sepeda yang terdaftar (registered) secara serupa dengan penyewa kasual.

* **total_count berkorelasi kuat dengan temp, atemp, casual, dan registered (0.63, 0.63, 0.67, dan 0.95)**. Ini berarti jumlah total penyewaan sepeda sangat dipengaruhi oleh suhu, penyewa kasual, dan penyewa terdaftar. Faktor kelembapan (hum) memiliki korelasi yang sedikit negatif dengan total penyewaan (-0.10).

Dari analisis ini, kita dapat menyimpulkan bahwa cuaca, terutama suhu, memiliki pengaruh signifikan terhadap jumlah penyewaan sepeda, baik dari penyewa kasual maupun yang terdaftar.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah terdapat perbedaan yang signifikan dalam pola penyewaan sepeda antara hari kerja dan akhir pekan?

```python
sns.barplot(x='workingday', y='total_count', data=day_df, estimator='mean')

plt.title('Perbandingan Rata-rata Penyewaan Sepeda Antara Hari Kerja dan Akhir Pekan')
plt.xlabel('Hari Kerja (1 = Hari Kerja, 0 = Akhir Pekan)')
plt.ylabel('Rata-rata Jumlah Penyewaan Sepeda')
plt.show()
```

Grafik menunjukkan bahwa rata-rata penyewaan sepeda sedikit lebih tinggi pada hari kerja dibandingkan akhir pekan, dengan penyewaan pada hari kerja di atas 4500 dan akhir pekan sedikit di bawahnya, kemungkinan karena penggunaan sepeda yang lebih banyak untuk keperluan rutin seperti bekerja atau sekolah pada hari kerja.

### Pertanyaan 2: Bagaimana pola penyewaan sepeda bervariasi berdasarkan faktor lingkungan dan musiman, seperti cuaca, musim, dan hari dalam seminggu?

```python
sns.lineplot(x='weekday', y='total_count', hue='season', data=day_df)

plt.title('Tren Penyewaan Sepeda Berdasarkan Musim dan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu (0 = Minggu, 6 = Sabtu)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.show()
```

Grafik menunjukkan bahwa jumlah penyewaan sepeda lebih tinggi pada musim panas dan semi dibandingkan musim gugur dan dingin, di mana musim dingin memiliki penyewaan terendah. Tren penyewaan juga lebih tinggi pada hari kerja (Senin hingga Jumat) dibandingkan akhir pekan, dengan lonjakan tajam pada hari Selasa di semua musim. Musim panas memiliki puncak penyewaan yang lebih stabil sepanjang minggu, sementara musim dingin menunjukkan jumlah penyewaan yang lebih rendah dan fluktuatif.
